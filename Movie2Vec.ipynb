```python
import pandas as pd
```

```python
tags = pd.read_csv("tags.csv")
tags
```

```python
tags_grouped = tags.groupby("movieId").aggregate(lambda tdf: tdf.values.tolist()).reset_index()
tags_grouped
```

```python
tags_grouped = tags_grouped[["tag", "movieId"]]
```

```python
tags_grouped
```

```python
movies = pd.read_csv("movies.csv")
```

```python
movies
```

```python
movies["genres"] = movies["genres"].apply(lambda x: x.split("|"))
movies
```

```python
movies["title"] = movies["title"].apply(lambda x: x.split())
movies
```

```python
movies["combined_title_genre"] = movies["title"].apply(lambda x: x[:-1])+movies["genres"]
movies
```

```python
movies = movies[["movieId", "combined_title_genre"]]
```

```python
movies["combined_title_genre"] = movies["combined_title_genre"].apply(lambda x:" ".join(x))
```

```python
movies
```

```python
tags_grouped["tag"] = tags_grouped["tag"].apply(lambda x: " ".join(x))
tags_grouped
```

```python
tag_genre = pd.merge(movies, tags_grouped, on=["movieId"], how="left")
tag_genre
```

```python
tag_genre["tag"] = tag_genre["tag"].fillna("")
```

```python
(tag_genre["combined_title_genre"] + " " + tag_genre["tag"])[0]
```

```python
tag_genre["sentences"] = tag_genre["combined_title_genre"] + " " + tag_genre["tag"]
```

```python
# sentence = tag_genre.loc[9741, "sentences"]
sentence = "Andrew Dice Clay:./.;[@%#$%^7] Dice Rules Comedy"
```

```python
import re
re.sub("\s{1,}", " ", re.sub("[^a-z0-9]", " ", sentence,0, flags=re.IGNORECASE), 0)
```

```python
tag_genre["sentences"] = tag_genre["sentences"].apply(lambda x: re.sub("\s{1,}", " ", re.sub("[^a-z0-9]", " ", x,0, flags=re.IGNORECASE), 0))
```

```python
tag_genre = tag_genre[["movieId", "sentences"]]
```

```python
tag_genre
```

```python
from gensim.models.word2vec import Word2Vec
import nltk
nltk.download('punkt')
```

```python
sentences = tag_genre["sentences"].tolist()
sentences = [nltk.word_tokenize(word) for sentence in sentences for word in nltk.sent_tokenize(sentence)]
```

```python
w2v = Word2Vec(sentences)
```

```python
len(w2v.wv.vocab.keys())
```

```python
w2v.save("w2v_model")
```

```python
sentences[0]
```

```python
tag_genre["sentence_token"] = ""
# tag_genre["sentence_token"] = tag_genre["sentence_token"].astype("object")
for i in range(len(sentences)):
  tag_genre.at[i, "sentence_token"] = sentences[i]
```

```python
tag_genre.loc[0, "sentences"]
```

```python
tag_genre
```

```python
movie_index_map = {mov_id: i+1 for i, mov_id in enumerate(tag_genre["movieId"].tolist())}
```

```python
import pickle
pickle.dump(movie_index_map, open("movie_index.pkl", "wb"))
```

```python
tag_genre.to_csv("tag_genre.csv")
```

```python
import numpy as np
movie_vector_matrix = np.empty((len(tag_genre["movieId"].tolist()), 100))
```

```python
model_keys = w2v.wv.vocab.keys()
```

```python
for row_num, row in tag_genre.iterrows():
  tag_vector = np.empty((len(row["sentence_token"]), 100))
  for i in range(len(row["sentence_token"])):
    if row["sentence_token"][i] in w2v.wv:
      tag_vector[i] = w2v.wv[row["sentence_token"][i]]
    else:
      tag_vector[i] = np.zeros((100))
  movie_vector_matrix[row_num] = np.mean(tag_vector, axis=0)
```
